```python
import pandas as pd
from datetime import datetime
```

```python
df_S08_SG1 = pd.DataFrame(episodes_S08_SG1)
df_S08_SG1.astype({'audience_us':'string', 'date_de_diffusion':'string'})
df_S08_SG1
```

```python
df_S08_SG1['audience_us'] = df_S08_SG1['audience_us'].apply(lambda x: x.replace(' millions', ''))
df_S08_SG1['audience_us'] = df_S08_SG1['audience_us'].apply(lambda x: x.replace(',', '.'))
df_S08_SG1['date_de_diffusion'] = df_S08_SG1['date_de_diffusion'].apply(lambda x: x.replace('\n', ''))
df_S08_SG1.astype({'audience_us':'float'})
df_S08_SG1
```

```python
month_list = {
    'janvier' : 1,
    'février' : 2,
    'mars' : 3,
    'avril' : 4,
    'mai' : 5,
    'juin' : 6,
    'juillet' : 7,
    'août' : 8,
    'septembre' : 9,
    'octobre' : 10,
    'novembre' : 11,
    'décembre' : 12}

df_S08_SG1['day'] = df_S08_SG1['date_de_diffusion'].apply(lambda x: x.split(' ')[0])
df_S08_SG1['month'] = df_S08_SG1['date_de_diffusion'].apply(lambda x: x.split(' ')[1]).map(month_list)
df_S08_SG1['year'] = df_S08_SG1['date_de_diffusion'].apply(lambda x: x.split(' ')[2])

df_S08_SG1
```

```python
df_S08_SG1['date_de_diffusion'] = pd.to_datetime(df_S08_SG1[['year', 'month', 'day']])
df_S08_SG1
```
