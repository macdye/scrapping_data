```python
import requests
import bs4
import re
import pandas as pd
```

```python
from scrap_S03_SGA import episodes_S03_SGA

df = pd.DataFrame(episodes_S03_SGA)
data = df.astype({'date de diffusion': 'string'})

data.replace('(\n)','',regex=True, inplace = True)
data.replace('( millions)','',regex=True, inplace = True)
data['audience_us'].replace('(,)','.',regex=True, inplace = True)
data['audience_fr'].replace('(,)','.',regex=True, inplace = True)
data.astype({'audience_us':'float'})
data
```

```python
month_list = {
    'janvier' : 1,
    'février' : 2,
    'mars' : 3,
    'avril' : 4,
    'mai' : 5,
    'juin' : 6,
    'juillet' : 7,
    'août' : 8,
    'septembre' : 9,
    'octobre' : 10,
    'novembre' : 11,
    'décembre' : 12
}

data['day'] = data['date de diffusion'].apply(lambda x: x.split(' ')[0]).apply(lambda x: x.replace('\n', ''))
data['month'] = data['date de diffusion'].apply(lambda x: x.split(' ')[1]).map(month_list)
data['year'] = data['date de diffusion'].apply(lambda x: x.split(' ')[2])
data['date de diffusion'] = pd.to_datetime(data[['year', 'month', 'day']])
data
```
